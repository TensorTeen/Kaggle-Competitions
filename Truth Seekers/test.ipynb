```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame()
```

```python
data['Strut No.'] = [1,2,3,4]
data['Strut Length(mm)'] = [370,470,470,520]
data['Strut Thickness(mm)'] = [1.0,1.0,1.0,1.0]
data['Strut Breadth(mm)'] = [30,30.3,24.4,30.4]
data['Ixx'] = (data['Strut Thickness(mm)']**3*data['Strut Breadth(mm)'])/12
data['Iyy'] = (data['Strut Breadth(mm)']**3*data['Strut Thickness(mm)'])/12
data['Buckling Load(N)_old'] = [25,20,8,17]
data['Buckling Load(N)'] = [28,20,15,18]
data['Buckling Load(N) Theory'] = data.apply(lambda row: (np.pi**2*210*row['Ixx'])/(row['Strut Length(mm)'])**2*1000, axis=1)

data['Error'] = (data['Buckling Load(N) Theory'] - data['Buckling Load(N)'])/data['Buckling Load(N) Theory']*100
data.drop(['Buckling Load(N)_old'], axis=1, inplace=True)
```

```python
data2 = pd.DataFrame()
```

```python
data2['Strut No.'] = [1,2,3,4]
data2['Strut Length(mm)'] = [360,460,460,510]
data2['Strut Thickness(mm)'] = [1.0,1.0,1.0,1.0]
data2['Strut Breadth(mm)'] = [30.3,30.3,24.4,30.4]
data2['Ixx'] = (data2['Strut Thickness(mm)']**3*data2['Strut Breadth(mm)'])/12
data2['Iyy'] = (data2['Strut Breadth(mm)']**3*data2['Strut Thickness(mm)'])/12
data2['Buckling Load(N)_old'] = [64,43,20,27]
data2['Buckling Load(N)'] = [68,40,28,32]
data2['Buckling Load(N) Theory'] = data2.apply(lambda row: (np.pi**2*210*2*row['Ixx'])/(row['Strut Length(mm)'])**2*1000, axis=1)
data2['Error'] = (data2['Buckling Load(N) Theory'] - data2['Buckling Load(N)'])/data2['Buckling Load(N) Theory']*100
data2.drop(['Buckling Load(N)_old'], axis=1, inplace=True)
```

```python
data2
```

```python
data2 = data2.drop(2)
data = data.drop(2)
```

```python
plt.plot(data2['Strut Length(mm)'],
         data2['Buckling Load(N) Theory'],
         label='Theory')
plt.plot(data2['Strut Length(mm)'],
         data2['Buckling Load(N)'],
         label='Practical')
plt.xlabel('Strut Length(mm)')
plt.ylabel('Buckling Load(N)')
plt.title('Buckling Load vs Strut Length For Fixed-Hinged Boundary Condition')
plt.legend()
plt.xlim(300,550)
plt.ylim(20,80)
```
